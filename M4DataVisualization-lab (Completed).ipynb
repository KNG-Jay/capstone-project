## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
QUERY = """
SELECT ConvertedComp FROM master
"""

cc = pd.read_sql_query(QUERY,conn)
cc.plot(kind='hist')
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """
SELECT Age FROM master
"""

age = pd.read_sql_query(QUERY,conn)
age.plot(kind='box')
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs FROM master
"""
aw = pd.read_sql_query(QUERY,conn)
aw.plot(x='Age', y='WorkWeekHrs', kind='scatter')
aw
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age FROM master
"""
wca = pd.read_sql_query(QUERY, conn)

ax0 = wca.plot(kind='scatter',
               x='WorkWeekHrs',
               y='Age',
               figsize=(14, 8),
               alpha=0.5,
               s=age
               )

ax1 = wca.plot(kind='scatter',
               x='CodeRevHrs',
               y='Age',
               alpha=0.5,
               color='purple',
               s=age,
               ax=ax0
               )

ax0.set_ylabel('Age')
ax0.set_xlabel('Hours')
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = """
SELECT DatabaseDesireNextYear,COUNT(DatabaseDesireNextYear) as Entries FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY COUNT(DatabaseDesireNextYear) Desc
LIMIT 5
"""
ddny = pd.read_sql_query(QUERY,conn)
ddny.set_index('DatabaseDesireNextYear', inplace=True)

ddny.plot(x='DatabaseDesireNextYear',
          y='Entries',
          kind='pie',
          figsize=(7, 8),
          autopct='%1.1f%%',
          xlabel='Database Desire Next Year')


          
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT Age, AVG(WorkWeekHrs) as AverageWorkHours, AVG(CodeRevHrs) as AverageReviewHours FROM master
WHERE AGE >=30 and Age <=35
GROUP BY Age
"""
wwcr = pd.read_sql_query(QUERY,conn)
wwcr.set_index('Age')

wwcr.plot.bar(x='Age', stacked=True)

```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT Age, AVG(ConvertedComp) as Median FROM master
WHERE Age >=45 and Age <=60
GROUP BY Age
"""
medcc = pd.read_sql_query(QUERY,conn)

medcc.plot(x='Age', kind='line')
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
SELECT MainBranch, COUNT(MainBranch) FROM master
GROUP BY MainBranch
"""
mb = pd.read_sql_query(QUERY, conn)
mb.set_index('MainBranch')

mb.plot(x='MainBranch', kind='barh')

```
